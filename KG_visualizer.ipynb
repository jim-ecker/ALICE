```python
import kuzu
import networkx as nx
from pyvis.network import Network

conn = kuzu.Connection(kuzu.Database("chat.db"))
result = conn.execute("""
    MATCH (e1:Entity)-[r:RELATES_TO]->(e2:Entity)
    RETURN e1.name, r.relation, e2.name, r.certainty_score
    LIMIT 500
""")

G = nx.DiGraph()
while result.has_next():
   row = result.get_next()
   G.add_edge(row[0], row[2], title=row[1], value=row[3])

net = Network(height="800px", notebook=True, directed=True)
net.from_nx(G)
net.show("graph.html")
```
